```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["chennai","lucknow","ahmedabad"],
    "temperature": [34,30,28],
    "humidity": [84,76,70]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new jersey","michigan","connecticut"],
    "temperature": [24,18,30],
    "humidity": [64,60,76]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
df.loc['us']
```

```python
windspeed_df = pd.DataFrame({
    "city": ["new jersey","michigan","connecticut"],
    "windspeed":[7,4,9]
},index=[0,1,2])

windspeed_df
```

```python
temperature_df = pd.DataFrame({
    "city": ["michigan","new jersey"],
    "temperature":[27,19]
},index=[1,0])

temperature_df
```

```python
pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
df1 = pd.DataFrame({
    "city": ["illionis","arizona","california","kansas","dallas"],
    "temperature":[23,29,24,19,32]
})
df1
                
```

```python
df2 = pd.DataFrame({
    "city": ['kansas','arizona','san diego'],
    "humidity": [58,67,70]
})
df2
```

```python
pd.merge(df1,df2, on='city')
```

```python
pd.merge(df1,df2, how="inner")
```

```python
pd.merge(df1,df2, how="left")
```

```python
pd.merge(df1,df2, how="right")
```

```python
pd.merge(df1,df2, how="outer")
```
