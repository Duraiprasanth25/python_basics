```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])

df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace=True)
df
```

```python
df.fillna(0)
df
```

```python

df.fillna(0) #(inplace=True) #inplace=True arguement is important because it will not modify dataframe but will create new dataframe orelse it will modify and save
df
```

```python
df.fillna({
    'temperature':df.temperature.mean(),
    'windspeed':df.windspeed.mean(),
    'event':'No Event'
})
```
