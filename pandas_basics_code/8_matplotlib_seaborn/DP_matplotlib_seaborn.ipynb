```python
import pandas as pd
df_sales = pd.read_excel('linechart.xlsx')
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.show() #plotting graph for single
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.show() #plotting graph for single by changing size
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"])
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"])
plt.show() #plotting graph for many
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color = "blue")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"], color = "green")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"], color = "red")
plt.show() #adding color
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color = "blue")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"], color = "green")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"], color = "red")

plt.title("product_sales")
plt.ylabel("Revenue")
plt.xlabel("Quarter")

plt.show() #adding label
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color = "blue", label = "Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"], color = "green",label = "Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"], color = "red",label = "Washing Machine")

plt.title("product_sales")
plt.ylabel("Revenue")
plt.xlabel("Quarter")
plt.legend()

plt.show() #adding legend
```

__PIE CHART__


```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales,labels = total_sales.index)
plt.show()
```

```python
plt.pie(total_sales,labels = total_sales.index,autopct="%1.1f%%")
plt.show()
```

```python
plt.pie(total_sales,labels = total_sales.index,autopct="%1.1f%%", explode=(0.2,0,0), shadow= True)
plt.show()
```

```python
plt.pie(total_sales,labels = total_sales.index,autopct="%1.1f%%", explode=(0.2,0,0), shadow= True, startangle = 180)
plt.show()
```
